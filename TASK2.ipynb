### **K-Means Clustering**

#### **K-means clustering is one of the popular unsupervised machine learning algorithms. The K-means algorithm identifies k number of centroids, and then allocates every data point to the nearest cluster, while keeping the centroids as small as possible.**

```python
# Predicting the optimum number of clusters for k-means classification
x = iris_dataframe.iloc[:, [0, 1, 2, 3]].values
from sklearn.cluster import KMeans
# Within Cluster Sum of Squares 
wcss = []
# Considering the clusters ranging from 1 to 10
for i in range(1, 11):
    k_means = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    k_means.fit(x)
    wcss.append(k_means.inertia_)
```

#### **Here, we use the Elbow method which is an empirical method to find the optimal number of clusters for a dataset. In this method, we pick a range of candidate values of k, then apply K-Means clustering using each of the values of k and then we find the average distance of each point in a cluster to its centroid, and represent it in a plot.**

```python
# Plotting the line graph
plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within Cluster Sum of Squares
plt.show()
```

#### **From the above graph of the 'elbow method', the optimum clusters is where the elbow occurs. This is when the within cluster sum of squares (WCSS) doesn't decrease significantly with every iteration. In this case the number of clusters is taken as 3.**

```python
# Applying kmeans to the dataset where the number of clusters is 3
k_means = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y = k_means.fit_predict(x)
```

```python
# Cluster centers
k_means.cluster_centers_
```

```python
# Cluster Visualization
# Visualizing the clusters on first two columns
plt.figure(figsize=(10,8))
plt.scatter(x[y == 0, 0], x[y == 0, 1], s = 100, c = 'pink', label = 'Iris-setosa')
plt.scatter(x[y == 1, 0], x[y == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y == 2, 0], x[y == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
# Plotting the centroids of the clusters
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:,1], s = 300, c = 'yellow', label = 'Centroids')
plt.legend()
```
